```python
from dataclasses import dataclass
@dataclass
class Schools:
    name:str 
    
    def print_name(self):
        print(self.name)
    def __post_init__(self):
        if not isinstance(self.name, str):
            raise TypeError(f"`name` must be a str, got {type(self.name).__name__!r} instead.")
        if not self.name.strip():
                raise ValueError("Agent name cannot be an empty or whitespace-only string.")

govt_schools = Schools(123)   
govt_schools.print_name()
```
