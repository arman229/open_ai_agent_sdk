```python
from agents import Agent,Runner,AsyncOpenAI,OpenAIChatCompletionsModel
import os  

try:
        
    GEMINI_API_KEY = os.getenv('GEMINI_API_KEY')
    if not GEMINI_API_KEY:
        raise ValueError('Gemini api key is not set in .env')
    client = AsyncOpenAI(base_url = 'https://generativelanguage.googleapis.com/v1beta/openai/',api_key=GEMINI_API_KEY)
    model = OpenAIChatCompletionsModel(model='gemini-1.5-flash',openai_client=client)
except Exception as e:
    print('Error',e)    
```

```python

basic_chatbot = Agent(name='basic_chatbot',instructions='You are helpful assistant ',model=model)
async def main(user_input:str):
   resp = await Runner.run(starting_agent=basic_chatbot,input=user_input)
   return resp.final_output
if __name__ == "__main__":
    final_resp = asyncio.run(main('what is AI'))
    print(final_resp)
    
```

```python

pak_agent = Agent(name='pak agent',instructions='You have knowledge an pakistant',model=model)
india_agent = Agent(name='india agent',instructions='You have knowledge an india',model=model)
iran_agent = Agent(name='iran agent',instructions='You have knowledge an iran',model =model)
triage_agent = Agent(name='triage agent',handoffs=[pak_agent,india_agent,iran_agent],instructions='You determine which agent to use based on the user question if you don\'t found any agent simple say i don\'t know ',model=model)

async def main(user_input:str):
   resp = await Runner.run(starting_agent=triage_agent,input=user_input)
   return resp.final_output
if __name__ == "__main__":
    final_resp = asyncio.run(main('pakistan history history'))
    print(final_resp)
    
```
