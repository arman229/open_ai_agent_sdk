````python
from pydantic import BaseModel, field_validator

class Product(BaseModel):
    name: str
    price: float

    @field_validator('name')
    def no_special_dcharacters(cls, v):
        if any(not c.isalnum() for c in v):
            raise ValueError("Name must contain only letters and numbers")
        return v
new_product = Product(name='ee3',price='12')
print(new_product)
````

````python
from datetime import datetime
from pydantic import BaseModel
import time
class EventBad(BaseModel):
    # This is evaluated **once**, when Python reads the class
    created_at: datetime = datetime.utcnow()

# Create two instances at different times
e1 = EventBad()
# wait a few secondsâ€¦
time.sleep(2)
e2 = EventBad()

print("e1.created_at:", e1.created_at)
print("e2.created_at:", e2.created_at)
# Both will show the SAME timestampâ€”even though e2 was created later!
````

default_factory=datetime.utcnow tells Pydantic to call the function every time a new model is instantiated.

````python
from datetime import datetime
from pydantic import BaseModel, Field

class EventGood(BaseModel):
    # datetime.utcnow is called **each time** you create an instance
    created_at: datetime = Field(default_factory=datetime.utcnow)

# Create two instances at different times
g1 = EventGood()
# wait a few secondsâ€¦
g2 = EventGood()

print("g1.created_at:", g1.created_at)  # e.g. 2025-06-24 12:00:00
print("g2.created_at:", g2.created_at)  # e.g. 2025-06-24 12:00:05 (later!)
````

````python
import random

def top_p_sampling(word_probs, top_p):
    # Step 1: Sort words by probability (descending)
    sorted_words = sorted(word_probs.items(), key=lambda x: x[1], reverse=True)

    # Step 2: Keep adding until cumulative probability >= top_p
    cumulative = 0.0
    selected_words = []
    for word, prob in sorted_words:
        cumulative += prob
        selected_words.append((word, prob))
        if cumulative >= top_p:
            break

    # Step 3: Normalize the selected probabilities
    total = sum(prob for _, prob in selected_words)
    normalized = [(word, prob / total) for word, prob in selected_words]

    # Step 4: Pick one word randomly
    words, probs = zip(*normalized)
    chosen_word = random.choices(words, weights=probs, k=1)[0]

    return chosen_word, selected_words

# Test with different top_p values
word_probs = {
    "blue": 0.6,
    "gray": 0.2,
    "pink": 0.1,
    "orange": 0.07,
    "green": 0.03
}

for top_p_value in [0.1, 0.6,0.7, 0.9, 1.0]:
    print(f"\nðŸŽ¯ top_p = {top_p_value}")
    chosen, selected = top_p_sampling(word_probs, top_p_value)
    print("Selected from:", [word for word, _ in selected])
    print("Final choice:", chosen)
````

Let's break down the `pre=True` parameter in Pydantic's `@validator` decorator to make it clear.

In Pydantic, when you define a custom validator using the `@validator` decorator, the `pre=True` parameter controls *when* the validator runs during the validation process. Here's a simple explanation:

### What Does `pre=True` Do?
- Normally, Pydantic validates fields by first *coercing* (converting) the input data to the expected type (e.g., converting a string `"123"` to an integer `123` if the field is typed as `int`).
- When you set `pre=True` in a `@validator`, the validator runs *before* this type coercion happens. This means your validator gets the *raw input data* as it was provided, not the converted data.

### Why Is This Useful?
Using `pre=True` is helpful when you need to validate or process the input data *before* Pydantic tries to convert it to the field's type. For example:
- You want to check if a string input matches a specific format (e.g., a phone number) before Pydantic tries to process it.
- You want to handle invalid inputs early and raise a custom error message.
- You need to transform the raw input before Pydantic's type conversion kicks in.

### Example to Clarify
Let's look at a practical example:

```python
from pydantic import BaseModel, validator

class MyModel(BaseModel):
    phone_number: str

    @validator("phone_number", pre=True)
    def check_phone_format(cls, value):
        # This runs BEFORE Pydantic tries to coerce the input to a string
        if not isinstance(value, str) or not value.startswith("+"):
            raise ValueError("Phone number must be a string starting with '+'")
        return value
```

#### Scenario 1: Without `pre=True` (Default Behavior)
If you remove `pre=True` (or set `pre=False`, which is the default), Pydantic will first try to coerce the input to the field's type (`str` in this case). For example:
- Input: `{"phone_number": 12345}`
- Pydantic coerces `12345` to `"12345"` *before* your validator runs.
- Your validator then sees `value` as the string `"12345"`, not the original integer `12345`.

This might be fine in some cases, but if you want to enforce that the input *must already be a string* and reject non-string inputs, you need `pre=True`.

#### Scenario 2: With `pre=True`
With `pre=True`, the validator runs on the raw input:
- Input: `{"phone_number": 12345}`
- Your validator sees `value` as `12345` (an integer, the raw input).
- The validator checks `isinstance(value, str)` and raises a `ValueError` because `12345` is not a string.
- Pydantic never gets to the type coercion step because the validator fails early.

For a valid input like `{"phone_number": "+1234567890"}`, the validator checks that itâ€™s a string starting with `+`, passes the validation, and then Pydantic proceeds with any further processing.

### Key Points
- **Default (`pre=False`)**: Validator runs *after* Pydantic's type coercion, so you get the coerced value (e.g., a string, integer, etc.).
- **`pre=True`**: Validator runs *before* type coercion, so you get the raw input (which could be any type, like a string, integer, dict, etc.).
- The validator still needs to return a value (or raise an error). The returned value is what Pydantic uses for further processing.
- Use `pre=True` when you need strict control over the raw input or want to validate its format/type before Pydanticâ€™s automatic conversions.

### Another Example (Real-World Use Case)
Suppose you have a field that should accept a string representing a date, but you want to validate its format before Pydantic tries to parse it:

```python
from pydantic import BaseModel, validator
import re

class Event(BaseModel):
    event_date: str

    @validator("event_date", pre=True)
    def validate_date_format(cls, value):
        if not isinstance(value, str):
            raise ValueError("event_date must be a string")
        if not re.match(r"^\d{4}-\d{2}-\d{2}$", value):
            raise ValueError("event_date must be in YYYY-MM-DD format")
        return value
```

- Input: `{"event_date": "2025-06-24"}` â†’ Passes (itâ€™s a string in the correct format).
- Input: `{"event_date": 20250624}` â†’ Fails (not a string, caught by `pre=True`).
- Input: `{"event_date": "24-06-2025"}` â†’ Fails (wrong format).

Without `pre=True`, Pydantic would coerce `20250624` to `"20250624"`, and youâ€™d need extra logic to catch that itâ€™s not in the right format.

### Gotchas
- If you use `pre=True`, be prepared to handle inputs of any type (e.g., strings, integers, lists, etc.), since no coercion has happened yet.
- The validator must still return a value that Pydantic can use for further validation or coercion.
- If you need to access other fieldsâ€™ values in a `pre=True` validator, use the `values` parameter (e.g., `def check_phone(cls, value, values)`), but note that other fields may not be validated or coerced yet.
