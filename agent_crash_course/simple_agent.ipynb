```python
import nest_asyncio
nest_asyncio.apply()
# 1. AGENT LEVEL 
import os
from agents import Agent,OpenAIChatCompletionsModel,AsyncOpenAI,Runner

try:
        
    GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
    if not GEMINI_API_KEY:
        raise ValueError("GEMINI ApI key not found")
    
    external_client = AsyncOpenAI(base_url='https://generativelanguage.googleapis.com/v1beta/openai/',api_key=GEMINI_API_KEY)
    model = OpenAIChatCompletionsModel(model='gemini-2.0-flash',openai_client=external_client)
    basic_agent = Agent(name='pak agent',instructions='You are a helpful assistant.',model = model)
    resp = await Runner.run(starting_agent=basic_agent,input= 'What is AI.')
    print(resp.final_output)
except Exception as e:
    print('Error',e)    
    
    
```
