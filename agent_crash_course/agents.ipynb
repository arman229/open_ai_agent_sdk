```python
import nest_asyncio

nest_asyncio.apply() 
import os
from agents import Agent, OpenAIChatCompletionsModel, AsyncOpenAI, Runner
import asyncio
try:

    GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
    if not GEMINI_API_KEY:
        raise ValueError("GEMINI ApI key not found")
  
    external_client = AsyncOpenAI(
        base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
        api_key=GEMINI_API_KEY,
    )
    model = OpenAIChatCompletionsModel(
        model="gemini-2.0-flash", openai_client=external_client
    )
except Exception as e:
    print("Error", e)
```

```python
def main(input):
    basic_agent = Agent(name='basic_agent',instructions='you are helpful assistant',model=model)
    resp = Runner.run_sync(basic_agent,input=input)
    print(resp.final_output)
if __name__=='__main__':
    main('what is ai')   
```

```python
async def main(input):
    basic_agent = Agent(name='basic_agent',instructions='you are helpful assistant',model=model)
    resp = await Runner.run(basic_agent,input=input)
    print(resp.final_output)
if __name__=='__main__':
    main('what is ai')   
```
