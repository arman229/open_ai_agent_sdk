# Generic means "flexible type."
It lets us write code that works with any data type without repeating the same code for each type.

üîÅ Instead of writing one class for int, one for str, one for dict, etc.,
we write one class using a placeholder type ‚Äî and Python fills in the real type later.

```python
from typing import TypeVar,Generic

T = TypeVar("T")
class Box(Generic[T]):
    def __init__(self,name:T):
        self.name=name
    def print_box(self):
        print(f"Type of name is {type(self.name)}")
        return self.name    

box =Box('square')  
print(box.print_box())      
        
```

```python
from typing import TypeVar,Generic

T1 = TypeVar("T1")
class Box1(Generic[T1]):
    def __init__(self,name:T1):
        self.name=name
    def print_box(self):
        print(f"Type of name is {type(self.name)}")
        return self.name    

box_1 =Box1(3)  
print(box_1.print_box())      
        
```

```python
from typing import TypeVar,Generic

T2 = TypeVar("T2")
class Box2(Generic[T2]):
    def __init__(self,name:T2):
        self.name=name
    def print_box(self):
        print(f"Type of name is {type(self.name)}")
        return self.name    

from typing import List

box_2: Box2[List[str]] = Box2(['square', 'circle'])  
print(box_2.print_box())   

   
        
        
```

```python
from typing import Union
# With Union[int, str], you are saying: ‚ÄúI only allow these two types.‚Äù
# With Generic[T], you're saying: ‚ÄúYou tell me the type, and I‚Äôll work with it.‚Äù
class UnionBox:
    def __init__(self, name: Union[int, str]):
        self.name = name
    def print_box(self):
        print(f"Type of name is {type(self.name)}")
        return self.name    
unionbox_1 = UnionBox(10)     
print(unionbox_1.print_box())   
unionbox_2 = UnionBox("hi")   
print(unionbox_2.print_box())   
# box3 = UnionBox([1, 2]) 
# print(box3.print_box())  
```
