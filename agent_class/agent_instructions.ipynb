```python
### Dynamic instructions

from agents import Agent, Runner,RunContextWrapper
from dataclasses import dataclass

@dataclass
class Question:
    lang:str

def dynamic_instruction(ctx:RunContextWrapper,agent:agent):
    resp = f'You are an helpful assistant. Response in this language :{ctx.context.lang}'
    return resp 

agent = Agent(
    name="basic agent",  
    instructions=dynamic_instruction,
    model=model,
)
user_query = Question('urdu')

resp = await Runner.run(starting_agent=agent, input="what is ai",context=user_query)
print(resp.final_output)
```
