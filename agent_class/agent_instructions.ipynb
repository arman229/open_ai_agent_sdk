```python
import asyncio
import time 

# Define async tool
@function_tool
async def async_fetch_data(query: str) -> str:
    start_time = time.time()
    print(f"[{query}] Start: {start_time:.2f}")
    await asyncio.sleep(5)  # Simulate delay
    end_time = time.time()
    print(f"[{query}] End: {end_time:.2f}")
    return f"Data for {query}"

# Create agent
agent = Agent(
    name="AsyncAgent",
    instructions="Use async_fetch_data to fetch data based on query.",
    tools=[async_fetch_data],model=model)

# Main async function
async def main():
    start = time.time()

    # Launch two tasks in parallel
    result1, result2 = await asyncio.gather(
        Runner.run_sync(agent, "weather"),
        Runner.run_sync(agent, "news")
    )

    end = time.time()

    print("\n--- Final Output ---")
    print("Weather:", result1.final_output)
    print("News:", result2.final_output)
    print(f"\nTotal Time: {end - start:.2f} seconds")

# Run it
asyncio.run(main())
```
